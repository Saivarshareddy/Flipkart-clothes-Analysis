```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time

Product_Name = []
Brand = []
Product_Type = []
Price = []
Original_price = []
Discount_percentage = []
Page_Num = []

total_time = time.time()

for i in range(1, 80): 
    start_time = time.time()
    URL =f"https://www.flipkart.com/search?q=clothes&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, "html.parser")

    for x in soup.find_all('div', attrs={'class':"p0C73x"}):
        pname = x.a
        brand = x.find('div', attrs={'class':"Fo1I0b"})  
        ptype = x.a
        cost = x.find('div', attrs={'class':"hZ3P6w"})
        op = x.find('div', attrs={'class':"kRYCnD"})
        discount = x.find('div', attrs={'class':"HQe8jr"})
        rating_div = x.find('div', class_=['MKiFS6', 'G_noJv'])
        
        rating = None 
        if rating_div:
            
            rating = rating_div.text.strip().replace('"', '')
        
        Product_Name.append(pname.text if pname else np.nan)
        Brand.append(brand.text if brand else np.nan)
        Product_Type.append(ptype.text if ptype else np.nan)
        Price.append(cost.text if cost else np.nan)
        Original_price.append(op.text if op else np.nan)
        Discount_percentage.append(discount.text if discount else np.nan)
        
        Page_Num.append(i)

    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time))
```

```python
import pandas as pd
df = pd.DataFrame({
    "Product_Name": Product_Name,
    "Brand": Brand,
    "Product_Type": Product_Type,
    "Price": Price,
    "Original_Price": Original_price,
    "Discount_Percentage": Discount_percentage,
    "Page_Num": Page_Num
})
```

```python
import pandas as pd
import numpy as np
import re

# --------------------------------------------------
# STEP 1: Check missing values
# --------------------------------------------------
print(df.isna().sum())


# --------------------------------------------------
# STEP 2: Remove rows with critical NULL values
# --------------------------------------------------
df = df.dropna(subset=[
    "Product_Name",
    "Price",
    "Original_Price"
])


# --------------------------------------------------
# STEP 3: Clean Price Columns
# --------------------------------------------------
def clean_price(value):
    if pd.isna(value):
        return np.nan
    value = re.sub(r"\D", "", str(value))
    return int(value) if value != "" else np.nan

df["Price_num"] = df["Price"].apply(clean_price)
df["Original_Price_num"] = df["Original_Price"].apply(clean_price)


# --------------------------------------------------
# STEP 4: Calculate Discount Amount
# --------------------------------------------------
df["Discount_Amount"] = df["Original_Price_num"] - df["Price_num"]


# --------------------------------------------------
# STEP 5: Extract Product Type from Product Name
# --------------------------------------------------
pattern = (
    r'(T[- ]?Shirt|Shirt|Kurta|Pant|Jeans|Dress|Top|Shorts|'
    r'Saree|Lehenga|Skirt|Gown|Salwar|Dupatta|Pyjama|Track Pant|Hoodie|Set)'
)

df["Product_Type"] = (
    df["Product_Name"]
    .astype(str)
    .str.extract(pattern, flags=re.IGNORECASE)[0]
    .str.title()
)


# --------------------------------------------------
# STEP 6: Handle remaining NaN Product Types
# --------------------------------------------------
df["Product_Type"] = df["Product_Type"].fillna("Other")


# --------------------------------------------------
# STEP 7: Final Cleaning (optional but recommended)
# --------------------------------------------------
df = df.dropna(subset=["Price_num", "Original_Price_num"])


# --------------------------------------------------
# STEP 8: Final verification
# --------------------------------------------------
print(df.isna().sum())
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=50,kde=True,color="green")
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

*The histogram shows that most products are priced between 300 and 400, with the highest concentration around 350, indicating this as the most common price range.

*The distribution is right-skewed, as a small number of high-priced products extend beyond 500 and reach up to around 1,200.

*This right tail increases the average price, even though the majority of products belong to the regular-price segment below 500.

*Overall, the graph indicates two pricing groups: a dominant mid-range segment and a smaller premium-priced segment.



```python
plt.figure(figsize=(10,6))
sns.histplot(df['Discount_Percentage'], bins=40, kde=True,color='blue')
plt.title('Distribution of Discount Percentage')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

*The distribution is slightly left-skewed, indicating fewer products with very low discounts.

*Only a small number of items have discounts below 40%, while very high discounts above 85% are rare.

*Overall, the graph suggests that heavy discounting is a common pricing strategy for these products.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Ratings', data=df, hue='Category', palette='viridis')
plt.title('Price vs. Ratings by Category', fontsize=14)
plt.xlabel('Price')
plt.ylabel('Ratings')
plt.legend(title='Category')
plt.show()
```

*The scatter plot shows no strong relationship between price and ratings, as products across all price ranges receive similar ratings.

*Most products are clustered in the 200500 price range, with ratings largely between 3.5 and 4.5.

*High-priced products do not consistently receive higher ratings compared to low- and medium-priced items.

*Ratings remain fairly stable across categories, suggesting that customer satisfaction is not heavily influenced by price.

*All three categories (Low, Medium, and High) have very similar average ratings, close to 4.0.

*The Medium category shows a slightly higher average rating, but the difference across categories is minimal.

*This indicates that customer satisfaction remains consistent regardless of price category.

*Overall, price category does not significantly impact average product ratings.
